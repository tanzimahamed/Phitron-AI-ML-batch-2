```python
try:
    # risky code
except:
    # error হলে এটা চলবে
else:
    # error না হলে এটা চলবে
```

```python
try:
    f = open("data.txt", "r")
    data = f.read()
except FileNotFoundError:
    print("File not found")
finally:
    print("Closing file")
```

# raise Error
raise ব্যবহার করে তুমি নিজে থেকে exception throw (error তৈরি) করতে পারো।

মানে:
Python নিজে error না দিলেও → তুমি চাইলে error তৈরি করতে পারো

```python
x = -5

if x < 0:
    raise ValueError("Negative number not allowed")
```

## Why use raise?
1. Invalid input আটকাতে
2. Security check করতে
3. Business logic enforce করতে

```python
try:
    age = int(input("Enter age: "))

    if age < 18:
        raise Exception("You are not eligible")

    print("Access granted")

except Exception as e:
    print(e)
```

```python
def login(password):
    if password != "1234":
        raise Exception("Wrong password")
    return "Login successful"

try:
    print(login("1111"))
except Exception as e:
    print(e)
```

## Custom Exception

 Python-এর built-in errors (ValueError, TypeError) ছাড়াও


তুমি চাইলে নিজের মতো error বানাতে পারো

```python
class MyError(Exception):
    pass

raise MyError("Something went wrong")
```

```python
class AgeError(Exception):
    pass

age = int(input("Enter age: "))

if age < 18:
    raise AgeError("You are not eligible")

print("Access granted")
```

```python
class AgeError(Exception):
    pass

try:
    age = int(input("Enter age: "))

    if age < 18:
        raise AgeError("You are not eligible")

    print("Access granted")

except AgeError as e:
    print("Error caught:", e)
```

```python
class InsufficientBalance(Exception):
    pass

balance = 1000
withdraw = int(input("Enter amount: "))

if withdraw > balance:
    raise InsufficientBalance("Not enough balance")

balance -= withdraw
print("Remaining balance:", balance)
```

```python
class AgeError(Exception):
    pass

def check_age(age):
    if age < 18:
        raise AgeError("Age must be 18+")
    return "Allowed"

try:
    print(check_age(15))
except AgeError as e:
    print("Error:", e)
```
