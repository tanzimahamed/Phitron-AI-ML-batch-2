## Univariate Analysis--Categorical value analysis



```python
# Categorical value analysis
sns.countplot(data= df, x=df['Survived'])
```

```python
df['Survived'].value_counts()
```

```python
#persentage
(df['Survived'].value_counts()/len(df))*100
```

```python
Survived_Count = df['Survived'].value_counts()


survived_label= df['Survived'].unique()
print(survived_label)

plt.figure(figsize=(8,8))
plt.pie(Survived_Count,labels=survived_label, autopct="%1.1f" )
plt.title('Pie chart of Survived rate')
plt.show()
```

## Univariate Analysis ----> Numerical Column Analysis

```python
sns.histplot(df['Age'], bins=50)

plt.title('Histrogram of Age ')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.kdeplot(df['Age'])

plt.title('KDE Plot of Age ')
plt.xlabel('Age')
plt.ylabel('probability distribution ')
plt.show()
```

```python
sns.histplot(df['Fare'], bins=60)

plt.title('Histrogram of Fare ')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.kdeplot(df['Fare'])

plt.title('KDE Plot of Fare ')
plt.xlabel('Fare')
plt.ylabel('probability distribution ')
plt.show()
```
