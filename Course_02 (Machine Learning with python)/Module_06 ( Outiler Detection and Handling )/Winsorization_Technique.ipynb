```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('marks_dataset.csv')
```

```python
sns.kdeplot(data = df , x='maths_marks')
```

```python
sns.boxplot(data =df , x = 'maths_marks')
```

```python
df['maths_marks'].describe()
```

```python
x = 12 # koto percentile dhore kaj kortisi
x = x/100
min_range = df['maths_marks'].quantile(x) # 0.1 -> 1 percentile
max_range = df['maths_marks'].quantile(1-x) # 1- 0.1 = 0.99 -> 99 percentile

print(f"min range : {min_range} and max range: {max_range}")
```

```python
df['maths_marks'] =df['maths_marks'].clip(min_range,max_range)
```
