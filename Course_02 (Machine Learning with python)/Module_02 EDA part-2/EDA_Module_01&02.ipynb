# Multi-variate Analysis
Countplot with hue

**Multivariate analysis** is a statistical technique used to analyze datasets containing more than two variables simultaneously to understand the complex relationships, patterns, and dependencies between them



Multivariate Analysis (মাল্টিভেরিয়েট অ্যানালাইসিস) হলো এমন এক ধরণের পরিসংখ্যানিক পদ্ধতি (Statistical Method) যা একই সময়ে দুই বা ততোধিক চলক (Variables) বা ফ্যাক্টরের মধ্যে সম্পর্ক এবং তাদের প্রভাব বিশ্লেষণ করে।

```python
sns.countplot(data= df, x=df['Sex'])
```

```python
sns.countplot(data= df, x=df['Sex'],hue=df['Survived'])
```

```python
gender_split=df.groupby('Sex')['Survived'].value_counts()
gender_split
```

```python
gender_split=df.groupby('Sex')['Survived'].value_counts(normalize=True)
gender_split
```

```python
sns.countplot(data= df, x=df['Pclass'],hue=df['Survived'])
```

```python
Pclass_split=df.groupby('Pclass')['Survived'].value_counts(normalize=True)
Pclass_split
```

DE (Kernel Density Estimation) প্লট হলো ডেটাসেটের মানগুলোর বিন্যাস বা ডিস্ট্রিবিউশন (Distribution) দেখার একটি জনপ্রিয় পদ্ধতি। এটি অনেকটা হিস্টোগ্রামের (Histogram) মতো, তবে বার (bar) বা খোপের পরিবর্তে এটি একটি মসৃণ ও নিরবচ্ছিন্ন রেখা (Smooth Curve) তৈরি করে ডেটার সম্ভাব্যতা ঘনত্ব (Probability Density) প্রকাশ করে

```python
sns.kdeplot(data=df,x='Age',hue='Survived')
```
