```python
First = np.random.randint(1,100, size=(2,3))
Scend = np.random.randint(10,70, size=(2,3))



print(First)
print(Scend)
```

Concatenate


```python
con_row = np.concatenate([First,Scend], axis=0)
print(con_row)    # row wise

print("_________________________________")

con_colum = np.concatenate([First,Scend], axis=1)
print(con_colum)   # column wise
```

Arithmetic + Math Operator


```python
# Faster for vectorization


x=np.array([10,20,30,40,50,60,70])
y=np.array([1,2,3,4,5,6,7])


add= np.add(x,y)
print(add)

print(x+y)  #Add


sub= np.subtract(x,y)
print(sub)

print(x-y)  #Sub



Mul= np.multiply(x,y)
print(Mul)

print(x*y)  #Mul
print(x/y)  #Div
print(x%y)  #Mod
print(x**y) #Power
```

Mathmatical function


```python
# Trigonometry

# Sine cos tan takes values in radine
sin_val= np.sin(x)
print(sin_val)

cos_val= np.cos(x)
print(cos_val)

tan_val= np.tan(x)
print(tan_val)
```

```python
# Radine to degree

deg_conversion= np.rad2deg(x)
print(deg_conversion)
```

```python
# Log

log_val= np.log(x)
print(log_val)


base_10= np.log10(x)
print(base_10)

base_2= np.log2(x)
print(base_2)
```

```python
import numpy as np


d = np.array([10,7,8,16,73,12,18,24])
f = np.array([1,2,3,4,5,6,7,8])

getter_then= d > f

print(getter_then)


equal = d==f
print(equal)




# all()---->sob true ki na
# any ()-----> jekono ekta true kina

print(getter_then.all())
print(getter_then.any())



```

Sorting


```python
# INplace


t= np.array([10,36,4,6,23,4,3,11,1,55,90])
z= t.copy()
print(z)

z.sort()
print(z)

print(t)
```

```python
# Copy Sorting

print(t)
sort_arr = np.sort(t)

print(sort_arr)

print (t)
```

```python
# 2D array sorting

mat= np.array([[33,22,11,44,66,64,32,],
               [77,55,44,33,232,556,35]
               ])


#Horizon axis=1

hor_sort= np.sort(mat,axis=1)
print(hor_sort)


# Vertical axis=0

ver_sort= np.sort(mat,axis=0)
print(ver_sort)
```
